```python
def a(x, y):
  result = x + y * 2
  return result
b = a(1,2)
c = a(2,3)
print(b, c)
```

```python
def a(x):
  result = 0
  while result < 10:
    result = result + x
    if result == 5:
      return result
  return result
b = a(1)
c = a(2)
print(b, c)
```

原因：<br/>b從0+1、1+1...加到5之後達到if迴圈的條件，所以return回5<br/>c從0+2、2+2...加到10後就不滿足while迴圈運行的條件，所以return回10

<h2>遞迴函式

```python
def tf(value):
  return 2 * value - 1
test_values = [0.0,0.0011,0.25,0.5,0,54,0.75,1.0]
scaled_values = [tf(value) for value in test_values]
print("Original values:",test_values)
print("Scaled values:",scaled_values)
```

```python
def a(n):
  if n == 0 or n == 1:
    return 1
  else: return n + a(n-1)
print(a(3))
```

a(3)因為3!=1，會進入else，return 3 + a(2)<br/>a(2)因為2!=1，進入else，return 2 + a(1)<br/>a(1)因為1==1，return 1，<br/>所以a(3) = 3 + a(2) = 3 + 2 + a(1) = 3 + 2 + 1 = 6<br/>同理，若為a(5)則a(5) = 5 + 4 + 3 + 2 + 1 = 15
